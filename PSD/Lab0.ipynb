### Exemplo 5 — Geração de um sinal senoidal
Um senoide discreto pode ser escrito como:

$$ x[n] = A\sin\left(2\pi f\frac{n}{f_s}\right) $$

onde:
- `A` é a amplitude;
- `f` é a frequência do sinal;
- `f_s` é a frequência de amostragem.


````python

fs = 8000          # frequência de amostragem (Hz)
duracao = 1.0      # segundos
f0 = 440           # frequência do senoide (Hz)
A = 0.8            # amplitude

n = np.arange(int(fs*duracao))
x = A*np.sin(2*np.pi*f0*n/fs)

print("Número de amostras:", len(x))
print("Valor mínimo:", x.min())
print("Valor máximo:", x.max())

plt.plot(n[:200], x[:200])
plt.title("Trecho inicial do sinal senoidal")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
plt.show()
````

### Exemplo 6 — Salvando e lendo um arquivo WAV
Arquivos WAV normalmente armazenam inteiros. Por isso, converteremos o sinal para `int16`.


````python

arquivo_wav = output_dir / "senoide_440Hz.wav"

x_int16 = np.int16(x * 32767)
wavfile.write(arquivo_wav, fs, x_int16)

fs_lido, x_lido = wavfile.read(arquivo_wav)

print("Arquivo salvo em:", arquivo_wav)
print("fs_lido =", fs_lido)
print("dtype lido =", x_lido.dtype)
print("shape lido =", x_lido.shape)
````

## Parte 3 — Sinais digitais 2D (imagens)

Uma imagem em tons de cinza pode ser representada por uma matriz 2D, em que cada elemento corresponde à intensidade de um pixel.


### Exemplo 8 — Criando uma imagem sintética
Vamos criar uma imagem 2D sem depender de arquivos externos.


````python

H, W = 256, 256
lin = np.linspace(0, 255, H, dtype=np.float32)
col = np.linspace(0, 255, W, dtype=np.float32)

imagem = 0.6*lin[:, None] + 0.4*col[None, :]
imagem = np.clip(imagem, 0, 255).astype(np.uint8)

plt.imshow(imagem, cmap="gray")
plt.title("Imagem sintética em tons de cinza")
plt.colorbar()
plt.show()

print("shape:", imagem.shape)
print("dtype:", imagem.dtype)
````

### Exemplo 9 — Salvando e lendo uma imagem


````python

arquivo_img = output_dir / "imagem_sintetica.png"
iio.imwrite(arquivo_img, imagem)

imagem_lida = iio.imread(arquivo_img)

print("Arquivo salvo em:", arquivo_img)
print("shape da imagem lida:", imagem_lida.shape)
print("dtype da imagem lida:", imagem_lida.dtype)

plt.imshow(imagem_lida, cmap="gray")
plt.title("Imagem lida do arquivo")
plt.axis("off")
plt.show()
````

### Exemplo 10 — Operações simples em imagens
Duas operações clássicas são:
- **negativo**: `255 - imagem`;
- **limiarização**: separar pixels acima e abaixo de um valor.


````python

negativo = 255 - imagem_lida
binaria = (imagem_lida >= 128).astype(np.uint8) * 255

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(imagem_lida, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(negativo, cmap="gray")
plt.title("Negativo")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(binaria, cmap="gray")
plt.title("Limiarizada")
plt.axis("off")

plt.tight_layout()
plt.show()
````
